### iii. Als het goed is, kom je met deze policy tot een onvolledige value function. Waarom is dat het geval? Hoe kan je dat oplossen?
In dit geval komen we niet op elke state en berekenen we dus ook geen value voor deze states, dit komt doordat we maar 1 episode bestaat die niet door elke state heen gaat.
Dit kunnen we oplosssen door bij elke state transitie een lage kans te nemen om naar een andere state toe te gaan dan dat de bedoeling was. Zolang we de berekeningen maar lang genoeg draaien komt het algoritme dan vaak genoeg op alle states terecht.

### iv. Stel, je maakt gebruik van any-visit Monte-Carlo prediction (in plaats van first-visit). Maakt dat een verschil voor het resultaat? Waarom wel/niet?
Bij any-visit maak je gebruik van elke keer dat je een state bezoekt om de waarde van de state te berekenen en bij first-visit maak je alleen gebruik van de eerste keer dat je de state bezoekt. Wanneer we gebruik maken van deze policy komen we niet meer dan 1 keer in dezelfde state terecht, er zal dan ook geen andere berekening gemaakt worden dan bij first-visit als we any-visit gebruiken.

## C. On-policy first-visit Monte-Carlo Control

### i. Leg in eigen woorden uit wat een Q-function is.
Een Q-functie vervangt als het ware het idee dat een state een lijst van values heeft voor een paar van 1 state en 1 action samen dat een lijst van values heeft.\
$\{state: [values]\}$ --> $\{(state, action): [values]\}$
